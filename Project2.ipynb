# Q4) Is there a correlation between the number of enrolled students and the number of passout students?

```python
correlation = data['No. of enrolled students'].corr(data['No. of Passout Students'])
print(f"Pearson correlation coefficient: {correlation}")

# Creating a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['No. of enrolled students'], data['No. of Passout Students'],color='blue')
plt.xlabel('Number of Enrolled Students')
plt.ylabel('Number of Passout Students')
plt.title('Correlation between Enrolled and Passout Students')
plt.show()
```

There is a strong positive correlation (approximately 1) between the number of enrolled students and the number of passout students.

# Q7) Which state/UT has the most significant gender disparity in passout rates?

```python
# Calculating the absolute difference in passout rates between males and females
data['Gender Disparity'] = abs(data['% of male passout'] - data['% of female passout'])
max_disparity = data[data['Gender Disparity'] == data['Gender Disparity'].max()]
state_max_disparity = max_disparity['States/UTs'].values[0]

# Plotting the gender disparity in each state/UT
plt.figure(figsize=(10, 6))
plt.bar(data['States/UTs'], data['Gender Disparity'], color='red')
plt.xlabel('States/UTs')
plt.ylabel('Gender Disparity (%)')
plt.title('Gender Disparity in Passout Rates in Each State/UT')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

state_max_disparity
```

# Q9) What is the average number of teachers per college across all states/UTs?

```python
# Calculating the average number of teachers per college
data['Teachers per College'] = data['No. of Teachers'] / data['No. of Colleges']
average_teachers_per_college = data['Teachers per College'].mean()

# Plotting the average number of teachers per college in each state/UT
plt.figure(figsize=(10, 6))
plt.bar(data['States/UTs'], data['Teachers per College'], color='orange')
plt.xlabel('States/UTs')
plt.ylabel('Teachers per College')
plt.title('Average Number of Teachers per College in Each State/UT')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

average_teachers_per_college
```

# Q10) Which state/UT has the highest number of colleges per capita (per 1,000 students)?

```python
# Calculating the number of colleges per capita (per 1,000 students)
data['Colleges per 1000 Students'] = data['No. of Colleges'] / (data['No. of enrolled students'] / 1000)
max_colleges_per_capita = data[data['Colleges per 1000 Students'] == data['Colleges per 1000 Students'].max()]
state_max_colleges_per_capita = max_colleges_per_capita['States/UTs'].values[0]

# Plotting colleges per 1000 students
plt.figure(figsize=(10, 6))
plt.bar(data['States/UTs'], data['Colleges per 1000 Students'], color='orange')
plt.xlabel('States/UTs')
plt.ylabel('Colleges per 1000 Students')
plt.title('Colleges per 1000 Students in Each State/UT')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

state_max_colleges_per_capita
```

# Q11) What is the ratio of male to female passout students across all states/UTs? 

```python
# Calculating the total number of male and female passout students
total_male_passout = (data['% of male passout'] / 100) * data['No. of Passout Students']
total_female_passout = (data['% of female passout'] / 100) * data['No. of Passout Students']
ratio_male_to_female = total_male_passout.sum() / total_female_passout.sum()

# Plotting the male to female passout ratio
plt.figure(figsize=(10, 6))
plt.bar(data['States/UTs'], total_male_passout, label='Male Passout Students', alpha=0.7)
plt.bar(data['States/UTs'], total_female_passout, label='Female Passout Students', alpha=0.7, bottom=total_male_passout)
plt.xlabel('States/UTs')
plt.ylabel('Number of Passout Students')
plt.title('Male and Female Passout Students in Each State/UT')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

ratio_male_to_female
```

# Q12) What is the percentage of states/UTs with a passout rate above 75%?

```python
# Calculating the percentage of states/UTs with passout rate above 75%
above_75_percent = data[data['% of Passout students'] > 75]
percentage_above_75 = (len(above_75_percent) / len(data)) * 100

# Plotting the distribution of passout rates
plt.figure(figsize=(10, 6))
plt.hist(data['% of Passout students'], bins=5, color='purple', edgecolor='black')
plt.axvline(x=75, color='red', linestyle='--', label='75% Threshold')
plt.xlabel('Passout Rate (%)')
plt.ylabel('Number of States/UTs')
plt.title('Distribution of Passout Rates')
plt.legend()
plt.tight_layout()
plt.show()

percentage_above_75
```

# Q13) What is the overall passout rate across all states/UTs?

```python
# Calculating the overall passout rate
total_enrolled_students = data['No. of enrolled students'].sum()
total_passout_students = data['No. of Passout Students'].sum()
overall_passout_rate = (total_passout_students / total_enrolled_students) * 100

# Plotting the passout rate for each state/UT
plt.figure(figsize=(10, 6))
plt.bar(data['States/UTs'], data['% of Passout students'], color='green')
plt.axhline(y=overall_passout_rate, color='red', linestyle='--', label=f'Overall Passout Rate: {overall_passout_rate:.2f}%')
plt.xlabel('States/UTs')
plt.ylabel('Passout Rate (%)')
plt.title('Passout Rate in Each State/UT')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

overall_passout_rate
```

# Q14) Which state/UT has the lowest % of teachers as Compared with No. of enrolled Students?

```python
min_teachers = data[data['No. of Teachers'] == data['No. of Teachers'].min()]
state_min_teachers = min_teachers['States/UTs'].values[0]

# Plotting the number of teachers in each state/UT
plt.figure(figsize=(10, 6))
plt.bar(data['States/UTs'], data['No. of Teachers'], color='skyblue')
plt.xlabel('States/UTs')
plt.ylabel('Number of Teachers')
plt.title('Number of Teachers in Each State/UT')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

state_min_teachers
```

# Q15) What is the percentage distribution of total enrolled students among the top 5 states/UTs with the highest enrollment?

```python
# Sorting the states/UTs by the number of enrolled students in descending order and selecting the top 5
top_5_enrollment = data.sort_values(by='No. of enrolled students', ascending=False).head(5)

# Calculating the total number of enrolled students in the top 5 states/UTs
top_5_total_enrollment = top_5_enrollment['No. of enrolled students'].sum()

# Calculating the percentage distribution of enrolled students among the top 5 states/UTs
top_5_enrollment['Enrollment Percentage'] = (top_5_enrollment['No. of enrolled students'] / top_5_total_enrollment) * 100

# Plotting the percentage distribution as a pie chart
plt.figure(figsize=(7, 7))
plt.pie(top_5_enrollment['Enrollment Percentage'], labels=top_5_enrollment['States/UTs'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Percentage Distribution of Total Enrolled Students Among Top 5 States/UTs')
plt.show()
```
